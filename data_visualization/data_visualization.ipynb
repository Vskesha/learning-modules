```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
date = pd.date_range(start = '2024-02-01', freq='D', periods=8)
plt.plot(date, [23, 17, 17, 16, 15, 14, 17, 20])
plt.show()
```

```python
date = pd.date_range(start = '2024-02-01', freq='D', periods=8)
plt.plot(date, [23, 17, 17, 16, 15, 14, 17, 20], label='day temperature')
plt.xlabel('', fontsize='small', color='midnightblue')
plt.ylabel('', fontsize='small', color='midnightblue')
plt.title('  ', fontsize=15)
plt.text(date[3], 20, '  ', color='blue')
plt.legend()
plt.show()
```

```python
date = pd.date_range(start = '2024-02-01', freq='D', periods=8)
fig, axs = plt.subplots()
axs.plot(date, [23, 17, 17, 16, 15, 14, 17, 20], label='day temperature')
axs.plot(date, [19, 11, 16, 11, 10, 10, 11, 16], label='night temperature')
plt.xlabel('', fontsize='small', color='midnightblue')
plt.ylabel('', fontsize='small', color='midnightblue')
plt.title('  ', fontsize=15)
plt.text(date[3], 20, '  ', color='blue')
plt.legend()
plt.show()
```

```python
date = pd.date_range(start = '2024-02-01', freq='D', periods=8)
fig, axs = plt.subplots(2, 1)
axs[0].plot(date, [23, 17, 17, 16, 15, 14, 17, 20], label='day temperature')
axs[1].plot(date, [19, 11, 16, 11, 10, 10, 11, 16], label='night temperature')
axs[0].set_title('', fontsize=10)
axs[1].set_title('', fontsize=10)
fig.suptitle('  ', fontsize=15)
plt.show()
```

```python
date = pd.date_range('2024-02-01', freq='D', periods=8)
plt.plot(
    date,
    [23, 17, 17, 16, 15, 14, 17, 20],
    label='day temperature',
    color='#FF5733',
    linestyle='--'
)
plt.plot(
    date,
    [19, 11, 16, 11, 10, 10, 11, 16],
    label='night temperature',
    color='#061358',
    linestyle='dashdot'
)
plt.xlabel('', fontsize='small', color='midnightblue')
plt.ylabel('', fontsize='small', color='midnightblue')
plt.title('  ', fontsize=15)
plt.legend()
plt.show()
```

```python
date = pd.date_range('2024-02-01', freq='D', periods=8)
plt.rc('font', size=10)
plt.rc('xtick', labelsize=6)
plt.plot(
    date,
    [23, 17, 17, 16, 15, 14, 17, 20],
    label='day temperature',
    color='#FF5733',
    linestyle='--',
    linewidth=2,
    marker='D'
)
plt.plot(
    date,
    [19, 11, 16, 11, 10, 10, 11, 16],
    label='night temperature',
    color='#061358',
    linestyle='dashdot',
    linewidth=2,
    marker='^'
)
plt.ylim(0, 25)
plt.xlabel('', fontsize='small', color='midnightblue')
plt.ylabel('', fontsize='small', color='midnightblue')
plt.title('  ', fontsize=15)
plt.legend()
plt.grid()
plt.show()
```

```python
fig, ax = plt.subplots()
ax.bar(
    ['', '', '', ' '],
    [39, 38, 27, 22],
    color=['b', 'r', 'y', 'g'],
    width=0.7,
    linewidth=3,
    edgecolor='black'
)
plt.ylim(0, 50)
plt.xlabel('', fontsize='small', color='midnightblue')
plt.ylabel('', fontsize='small', color='midnightblue')
plt.title(' :    2020', fontsize=15)
ax.set_axisbelow(True)
ax.grid(axis='y')
plt.show()
```
