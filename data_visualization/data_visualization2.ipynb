```python
import matplotlib.pyplot as plt

labels = [
    "Junior Software Engineer",
    "Senior Software Engineer",
    "Software Engineer",
    "System Architect",
    "Technical Lead"
]
data = [63, 31, 100, 2, 11]
explode = [0.15, 0, 0, 0, 0]
plt.pie(
    data,
    labels=labels,
    explode=explode,
    shadow=True,
    autopct="%.2f%%",
    pctdistance=1.15,
    labeldistance=1.35,
)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


theta = np.linspace(0, 2.0 * np.pi, 1000)

r = np.sin(theta * 6)
plt.polar(theta, r)
plt.show()
```
