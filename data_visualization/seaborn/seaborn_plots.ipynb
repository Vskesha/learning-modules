```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
data = sns.load_dataset('mpg')
data.head()
```

```python
sns.set_style('whitegrid')

date = pd.date_range(start='2024-03-05', freq='D', periods=8)
day = [23, 17, 17, 16, 15, 14, 17, 20]
night = [19, 11, 16, 11, 10, 10, 11, 16]
df = pd.DataFrame({
    'date': date,
    'day_temperature': day,
    'night_temperature': night,
})

sns.lineplot(data=df, markers=['D', '^'])
plt.show()
```

```python
sns.lineplot(data=df, x='night_temperature', y='day_temperature')
plt.show()
```

```python
sns.lineplot(data=df, x='night_temperature', y='day_temperature', sort=False)
plt.show()
```
