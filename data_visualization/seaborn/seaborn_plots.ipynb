```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
data = sns.load_dataset('mpg')
data.head()
```

```python
sns.set_style('whitegrid')

date = pd.date_range(start='2024-03-05', freq='D', periods=8)
day = [23, 17, 17, 16, 15, 14, 17, 20]
night = [19, 11, 16, 11, 10, 10, 11, 16]
df = pd.DataFrame({
    'date': date,
    'day_temperature': day,
    'night_temperature': night,
})

sns.lineplot(data=df, markers=['D', '^'])
plt.show()
```

```python
sns.lineplot(data=df, x='night_temperature', y='day_temperature')
plt.show()
```

```python
sns.lineplot(data=df, x='night_temperature', y='day_temperature', sort=False)
plt.show()
```

```python
sns.set_style('darkgrid')
data = sns.load_dataset('mpg')
sns.lineplot(data=data, x='model_year', y='horsepower', hue='origin')
plt.show()
```

```python
sns.scatterplot(data=data, x='mpg', y='displacement')
plt.show()
```

```python
sns.scatterplot(data=data, x='mpg', y='displacement', hue='origin')
plt.show()
```

```python
sns.relplot(x='mpg', y='displacement', kind='scatter', data=data)
plt.show()
```

```python
sns.relplot(data=data, x='mpg', y='displacement', kind='scatter', hue='origin', col='origin')
plt.show()
```
